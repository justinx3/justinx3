```python
x1=[6,8,9,3]
x2=[12,15,18,10,5]

y1=[7,10,4,5]
y2=[19,6,18,7,19]


#multiple plots in same graph

plt.plot(x1,y1)
plt.plot(x2,y2)
plt.show()
```

```python
plt.plot(x1,y1,c='g',label='plot1')
plt.plot(x2,y2,c='r',label='plot2')
plt.legend()
plt.show()
```

```python
data=pd.read_csv('auto-mpg.csv')
```

# HISTOGRAM

```python
plt.figure(figsize=(5,5))
plt.hist(data['mpg'])
plt.title('distribution of mpg')
plt.xlabel('MPG')
plt.ylabel('frequency')
plt.show()
```

```python
plt.figure(figsize=(4,4))
plt.hist(data['mpg'],color="red",rwidth=0.9)
plt.title('distribution of mpg')
plt.xlabel('MPG')
plt.ylabel('frequency')
plt.show()
```
